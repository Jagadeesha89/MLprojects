```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"D:\UJJIVANSFB.NS.csv")
```

```python
data1=data.drop(["Adj Close","Volume"], axis=1)
```

```python
plt.figure(figsize=(15,10))
plt.plot(data1["Close"])
plt.xlabel("Days")
plt.ylabel("Closeing Price")
```

```python
X=data1.drop(["Close","Date"],axis=1)
y=data1["Close"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=False)
Acc=[]
```

```python
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
Lr.fit(X_train,y_train)
```

```python
predict=Lr.predict(X_test)
```

```python
predict_df=pd.DataFrame({"Actula":y_test,"Predicted":predict})
```

```python
predict_df
```

```python
plt.figure(figsize=(15,8))
plt.plot(predict_df)
plt.legend(['Actula value','Predicted value'])
```

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(X_train,y_train)
```
